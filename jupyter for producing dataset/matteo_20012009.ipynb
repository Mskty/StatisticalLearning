```python
# Keeping 2001- 2009
dfnew = df.loc[(df.end_season<2010)].reset_index(drop=True)
dfnew.head()
```

### modifica di alcune colonne

```python
dfnew.drop(columns='Unnamed: 0',inplace=True)
```

```python
for col in dfnew.columns:
    print(col)
```

```python
dfnew.info()
```

```python
dfnew.seasons.head(5)
for i in range(0,5):
    print(len(dfnew.seasons[i]))
```

```python
# change seasons in the number of seasons (divide by 6 each string list lenght)
dfnew.seasons = dfnew.seasons.apply(len)
dfnew.seasons = dfnew.seasons/6
dfnew.seasons
```

```python
dfnew.head()
```

```python
# transform games_played and games started for the seasons
dfnew.games_played = dfnew.games_played / dfnew.seasons
dfnew.games_started = dfnew.games_started  / dfnew.seasons
dfnew.head()
```

```python
# drop index
dfnew.drop(columns='index',inplace=True)
```
